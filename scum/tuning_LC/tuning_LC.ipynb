```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.io
import scipy.stats
```

```python
data = {
    78: 718,
    66.5: 710,
    58: 702,
    36: 696,
    26: 690,
    18: 683,
    12: 677
}

temp = list(data.keys())
LSB = list(data.values())
```

```python
slope, intercept = scipy.stats.linregress(temp, LSB)[:2]
print('Slope:', slope, 'Intercept:', intercept)

x = np.arange(-10, 100)
plt.plot(temp[:3], LSB[:3])
plt.plot(temp[3:], LSB[3:])
# plt.plot(x, slope * x + intercept)
plt.title('LC_monotonic vs. temperature')
plt.xlabel('Measured temperature (C)')
plt.ylabel('LC_monotonic')
plt.show()
```

```python
data = {
    12.8: 196000,
    1.26: 180000,
    35: 164000,
    53: 176000,
    80: 192000
}

temp = list(data.keys())
frequency_per_LSB = list(data.values())
```

```python
slope, intercept = scipy.stats.linregress(temp, frequency_per_LSB)[:2]
print('Slope:', slope, 'Intercept:', intercept)

x = np.arange(-10, 100)
plt.plot(temp, frequency_per_LSB)
plt.plot(x, slope * x + intercept)
plt.show()
```
