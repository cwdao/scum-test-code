```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats
```

```python
coarse_code = [0, 8, 15, 20, 22, 24, 28, 30]
diff = {
    10: [0.00024, 0.00046, 0.00052, 0.00062, 0.00064, 0.00066, 0.00078, 0.00088],
    25: [0.0004, 0.0005, 0.00062, 0.00068, 0.00074, 0.00074, 0.00078, 0.00086],
    50: [0.00052, 0.00058, 0.0007, 0.00062, 0.00074, 0.00084, 0.00092, 0.00096],
    75: [0.00056, 0.00068, 0.00074, 0.00086, 0.0009, 0.00094, 0.00098, 0.00108],
}
```

```python
plt.plot(coarse_code, diff[10], label='10C')
plt.plot(coarse_code, diff[25], label='25C')
plt.plot(coarse_code, diff[50], label='50C')
plt.plot(coarse_code, diff[75], label='75C')
plt.title('LC frequency difference due to static divider')
plt.xlabel('LC coarse code')
plt.ylabel('LC frequency difference (GHz)')
plt.legend()
plt.show()
```

```python
x = np.arange(32)

slope, intercept = scipy.stats.linregress(coarse_code, diff[25])[:2]
print('Slope:', slope, 'Intercept:', intercept)
lin_approx = slope * x + intercept

plt.plot(coarse_code, diff[25])
plt.plot(x, lin_approx)
plt.title('LC frequency difference due to static divider at 25C')
plt.xlabel('LC coarse code')
plt.ylabel('LC frequency difference (GHz)')
plt.show()
```
