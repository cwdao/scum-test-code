```python
import numpy as np
import matplotlib.pyplot as plt
```

### Horizontal (x-axis)

```python
data_x = {
    0.5: -67,
    1: -73,
    2: -76,
    3: -80,
    4: -83,
    5: -87,
    5.5: -91,
    6: -94,
    6.5: -97,
    7: -99
}

distance_x = np.array(list(data_x.keys())) * 2.54
rssi_x = list(data_x.values())
```

```python
plt.figure(figsize=(8, 6))
plt.plot(distance_x, rssi_x)
plt.title('RSSI vs. distance along x-axis', fontsize=14)
plt.xlabel('Distance along x-axis (cm)', fontsize=14)
plt.ylabel('RSSI (dBm)', fontsize=14)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.savefig('pcb_rssi_x.pdf', bbox_inches='tight')
plt.show()
```
