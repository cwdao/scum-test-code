```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = {
    6: list(range(23, 32)),
    7: list(range(12, 32)),
    8: list(range(6, 28)),
    9: list(range(0, 15)),
    10: list(range(0, 9))
}

codes = np.array([[mid, fine] for mid in data.keys() for fine in data[mid]])
codes[:, 0]
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(codes[:, 0], codes[:, 1])
plt.title('Frequency tuning codes to receive BLE packets', fontsize=14)
plt.xlabel('Mid code', fontsize=14)
plt.ylabel('Fine code', fontsize=14)
plt.xticks([5, 6, 7, 8, 9, 10, 11], fontsize=14)
plt.yticks(np.arange(0, 32, 5), fontsize=14)
plt.text(7.8, 30, 'Board Q4 with coarse code=23', fontsize=14)
plt.savefig('ble_rx_codes.pdf', bbox_inches='tight')
plt.show()
```
