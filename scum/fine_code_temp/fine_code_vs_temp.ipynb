```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = {
    9: list(np.arange(0, 5)),
    20: list(np.arange(12, 21)),
    25: list(np.arange(18, 26)),
    31: list(np.arange(28, 31))
}

data_points = []

for temp in data.keys():
    data_points.extend([[temp, fine_code] for fine_code in data[temp]])

data_points = np.array(data_points)
```

```python
plt.figure(figsize=(10, 6))

plt.scatter(data_points[:, 0], data_points[:, 1], label="Fine codes received by phone")
temp = np.arange(7, 35)
plt.plot(temp, 1.222 * temp - 8.862, "--", c="C1", label="Linear fit")

plt.title('Fine code vs. temperature', fontsize=20)
plt.xlabel('Temperature ($^\circ$C)', fontsize=20)
plt.ylabel('Fine code', fontsize=20)
plt.xticks(fontsize=18)
plt.yticks(fontsize=18)
plt.legend(fontsize=18)
plt.text(21, 3, "coarse=22, mid=30", fontsize=18)
plt.savefig('fine_code_vs_temp.pdf', bbox_inches='tight')
plt.show()
```
